```python
def validation_isomap(data,nb_points,n_neighbors,noise):
    if data=="swiss":
        escargot, color = skl.datasets.make_swiss_roll(n_samples=nb_points, noise=noise)
    else:
        escargot, color = skl.datasets.make_s_curve(n_samples=nb_points, noise=noise)
        
    embedding = skl.manifold.Isomap(n_neighbors=n_neighbors, n_components=2)
    escargot_deroule = embedding.fit_transform(escargot)
    
    # Set up a figure twice as tall as it is wide
    fig = plt.figure(figsize=(14,5))
    fig.suptitle('Isomap')
    
    ax = fig.add_subplot(1, 2, 1, projection='3d')
    surf = ax.scatter(escargot[:,0], escargot[:,1], escargot[:,2], '.', c=color, cmap='jet')

    ax = fig.add_subplot(1, 2, 2)
    ax.scatter(escargot_deroule[:,0], escargot_deroule[:,1], c=color, marker='.', cmap='jet')
    plt.show()
```

```python
def validation_eigenmaps(data,nb_points,n_neighbors,noise):
    if data=="swiss":
        escargot, color = skl.datasets.make_swiss_roll(n_samples=nb_points, noise=noise)
    else:
        escargot, color = skl.datasets.make_s_curve(n_samples=nb_points, noise=noise)
        
    embedding = skl.manifold.SpectralEmbedding(n_neighbors=n_neighbors, n_components=2)
    escargot_deroule = embedding.fit_transform(escargot)
    
    # Set up a figure twice as tall as it is wide
    fig = plt.figure(figsize=(14,5))
    fig.suptitle('Isomap')
    
    ax = fig.add_subplot(1, 2, 1, projection='3d')
    surf = ax.scatter(escargot[:,0], escargot[:,1], escargot[:,2], '.', c=color, cmap='jet')

    ax = fig.add_subplot(1, 2, 2)
    ax.scatter(escargot_deroule[:,0], escargot_deroule[:,1], c=color, marker='.', cmap='jet')
    plt.show()
```

```python
validation_isomap("swiss",500,20,0.01)
```

```python
validation_eigenmaps("swiss",2000,20,0.01)
```

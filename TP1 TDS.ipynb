### Créer un signal u(t) impulsion périodique avec to= T/2 d'amplitude E
créer un vecteur t=[0, pas, T]

```python
T=20     #période
to=T/2 
E=10     #amplitude

t=np.arange(0, T, T/1000)

def signal(t):
    u=[]
    for i in range(len(t)):
        if np.pi * 2* np.sin(t[i])*(1/T) >0:
            u.append(E)
        else:
            u.append(-E)
    return(u)

plt.plot(t, E*np.sin(t))
plt.plot(t, signal(t))
plt.grid(True)
plt.show()
```
